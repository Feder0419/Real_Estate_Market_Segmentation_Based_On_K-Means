```python
import pandas as pd
from IPython.display import Image
from six import StringIO
import pydotplus
```

```python
dataset=pd.read_csv('C:/Users/tengd/OneDrive/0-论文/dataset.csv')
```

```python
range_x=dataset['WGS84X'].max()-dataset['WGS84X'].min()
range_y=dataset['WGS84Y'].max()-dataset['WGS84Y'].min()
print(range_x)
print(range_y)
```

```python
from sklearn.decomposition import PCA
data=pd.DataFrame(dataset,columns=['price202403','WGS84X','WGS84Y','county'])
# data['price202403']=data['price202403']*0.000006
countys=list(data['county'].unique())
data_list=[]
data_other=[]
for county in countys:
    temp=data.loc[data['county']==county]
    data2=temp.drop(columns=['county'])
    pca=PCA(n_components=2)
    data2=pca.fit_transform(data2)
    data_list.append(data2)
    data_other.append(temp)
```

```python
import numpy as np
import scipy.sparse
data_matrix_list=[]
for data in data_list:
    data_matrix=scipy.sparse.csr_matrix(data)
    data_matrix_list.append(data_matrix)
```

```python
from sklearn.cluster import KMeans
```

```python
for i in range(len(data_matrix_list)):
    data_matrix=data_matrix_list[i]
    label=KMeans(n_clusters=5,n_init=100).fit_predict(data_matrix)
    data_other[i]['label_1']=label
    print(data_other[i])
```

```python
import matplotlib.pyplot as plt
df=pd.concat(data_other,axis=0)
df.to_csv('C:/Users/tengd/OneDrive/0-论文/result_simpleKMeans_PCA.csv',index=False)
# data_matrix.shape
```
